```python
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
import numpy as np
import pandas as pd
import scipy
```

```python
def CntLags(df, column, n_lags=15):
    df[f'{column}_lag1'] = df[column]

    for lag in range(2, n_lags + 1):
        df[f'{column}_lag{lag}'] = [0] * df.shape[0]
    
    for lag in range(1, n_lags):
        df[f'{column}_lag{lag + 1}'] = [0] * lag + list(df[column][:-lag])
    
    if n_lags >= 5:
        df[f'{column}_mean5'] = np.mean([df[f'{column}_lag1'], df[f'{column}_lag2'], df[f'{column}_lag3'], df[f'{column}_lag4'], df[f'{column}_lag5']], axis = 0)
    if n_lags >= 10:
        df[f'{column}_mean10'] = np.mean([df[f'{column}_lag1'], df[f'{column}_lag2'], df[f'{column}_lag3'], df[f'{column}_lag4'], df[f'{column}_lag5'],
                                df[f'{column}_lag6'], df[f'{column}_lag7'], df[f'{column}_lag8'], df[f'{column}_lag9'], df[f'{column}_lag10'],], axis = 0)
    if n_lags >= 15:
        df[f'{column}_mean15'] = np.mean([df[f'{column}_lag1'], df[f'{column}_lag2'], df[f'{column}_lag3'], df[f'{column}_lag4'], df[f'{column}_lag5'],
                                df[f'{column}_lag6'], df[f'{column}_lag7'], df[f'{column}_lag8'], df[f'{column}_lag9'], df[f'{column}_lag10'],
                                df[f'{column}_lag11'], df[f'{column}_lag12'], df[f'{column}_lag13'], df[f'{column}_lag14'], df[f'{column}_lag15'],], axis = 0)

    return df
```

```python
blocks_info = pd.read_csv('blocks_info.csv')

blocks_info['time'] = pd.to_datetime(blocks_info['time'])
blocks_info = blocks_info.sort_values(by='time').reset_index()
blocks_info = blocks_info.drop(columns=['index', 'id'])
blocks_info
```

```python
df = pd.read_csv('binance_data.csv')
df['Open time'] = pd.to_datetime(df['Open time'])
df['Close time'] = pd.to_datetime(df['Close time'])
df = df.sort_values(by='Open time').reset_index()
df = df.drop(columns=['index'])
df
```

```python
blocks_info_columns = ['transaction_count', 'input_count', 'output_count', 'input_total_usd', 'fee_total_usd', 'fee_per_kb_usd', 'reward_usd', 'time_between_blocks']
```

```python
for column in blocks_info_columns:
    df[column] = [0]*df.shape[0]
```

```python
from tqdm import tqdm
```

```python
ind = 0
for i in tqdm(range(df.shape[0])):
    while df.loc[i, 'Open time'] > blocks_info.loc[ind, 'time']:
        ind += 1
    for column in blocks_info_columns:
        df.loc[i, column] = blocks_info.loc[ind, column]
```

```python
df['time_between_blocks'] = df['time_between_blocks'].fillna(df['time_between_blocks'].mean())
df
```

```python
df_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Quote asset volume', 'Number of trades', 'Taker buy base asset volume', 'Taker buy quote asset volume']
columns = df_columns + blocks_info_columns
```

```python
df_lookback = df.drop(columns=['Open time', 'Close time'])
orig_shape = df_lookback.shape

lookback = 15

lookback_columns = [df_lookback.shift(i) for i in range(lookback)]
df_lookback = pd.concat(lookback_columns, axis=1)

df_lookback = df_lookback.values.reshape(orig_shape[0], lookback, orig_shape[1])
df_lookback.shape, df.shape
```
